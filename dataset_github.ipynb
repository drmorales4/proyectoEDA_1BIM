## Limpieza de Datos

```python
# concatenar los dataframes
datatotal = pd.concat([data, data1, data2, data3, data4, data5, data6, data7, data8, data9, data10, data11, data12, data13, data14, data15, data16, data16, data17], axis=0)

# creamos un solo dataframe con todos los dataset que hemos leido.
```

```python
# Tipos de datos
datatotal.dtypes
```

```python
# Eliminar las columnas 
datatotal = datatotal.drop(['FIPS', 'Admin2', 'Province_State', 'Lat', 'Long_', 'Combined_Key'], axis=1)

# con el .drop eliminamos las columnas que son irrelevantes o no las vemos necesarias para el analisis
```

```python
# Inspeccionar
datatotal.head()
```

```python
# cambiar de nombre a las variables

datatotal = datatotal.rename(columns={'Country_Region':'country', 'Last_Update':'date'})
```

```python
datatotal.head() # inspeccionar
```

```python
# quitar la hora de la fecha
cont = 0

for x in datatotal['date']:
    valor = x.split(' ')[0]
    datatotal['date'].iloc[cont] = valor
    cont = cont + 1
```

```python
covidEcuador = datatotal[datatotal['country'] == "Ecuador"]  # Filtrado de la columna country

# generamos un nuevo dataframe solo con datos de Ecuador
```

```python
datatotal['date'] = pd.Series([datetime.strptime(d, "%Y-%m-%d") for d in datatotal["date"]])
```
